Get the "raw" data URL and import it into your `R` session using `read_csv`. Remember to capture the data under a variable name of your choosing!

```r
measles = read.csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-25/measles.csv')
measles |> head()
```

The `mmr` column has holds ü§í measles, mumps, rubella ü§í vaccination rates for students in each respective school. **If the `mmr` value is not available for a school the `mmr` value is set to -1 in this data set.** 

The target `mmr` vaccination rate as [recommended by the CDC](https://www.cdc.gov/mmwr/preview/mmwrhtml/mm5632a3.htm) for schools is 95%. 

**Calculate the fraction of schools per state that have vaccination rates greater than the CDC 95%. Capture the output as a table called `df_vacc_rates`.**

You'll need to do use `filter`, `group_by`, and `summarize`. 

I.e.
1. Filter out schools that don't report `mmr` vaccination rate (keep schools where `mmr >= 0`).
1. Group the data by `state`.
1. Summarize fracion of vaccination rates above 95%.

üí°Remember `n()` stands for the number or records in a group. Also, `sum(mmr >= 95, na.rm=TRUE)` will count the number of values above or equal to 95.üí°

```r
# create the df_vacc_rates here

df_vacc_rates = measles |>
                    filter(mmr >= 0)|>
                    group_by (state)|>
                    summarize (fraction_schools_per_state =  sum(mmr >= 95, na.rm=TRUE) /n())
df_vacc_rates
```

Which state (of those that report `mmr`) has the smallest fraction of schools above the CDC target vaccination rate of 95%?  

```r
df_vacc_rates |>
    top_n(n=1, wt = -fraction_schools_per_state) 
```

**Make a ECDF plot of the school vaccination rates in North Carolina.** Use the `overall` column which reports the "overall" vaccination rate at each school.

‚ùóÔ∏èRemember, you can calculate the `y` value for ECDF charts using `row_number`, `mutate`, and `arrange`.

```r
# Filter and arrange the data
nc_vacc_rates = measles |>
  filter(state == 'North Carolina', overall >= 0) |>
  arrange(overall)

# Calculate the fraction of schools values
ecdf_data = nc_vacc_rates |>
  mutate(fraction_of_schools = row_number() / n())

# Create the ECDF plot
options(repr.plot.res = 250, repr.plot.width = 5, repr.plot.height = 4)
ggplot(ecdf_data, aes(x = overall, y = fraction_of_schools)) +
  geom_step() +
  geom_point(shape = 1, size = 1, color = "darkgreen", fill = "transparent") + 
  labs(
    x = "Vaccination rate (%)",
    y = "Fraction of schools",
  ) + 
geom_vline(aes(xintercept=95), linetype = "dashed", color = 'blue') + 
annotate("text", x = 25, y = 0.85, label = "CDC recommended vaccine rate (95%)", hjust = -0.15, vjust = 2, color = 'blue') +
geom_curve(aes(x= 85, y=0.75, xend=95, yend=0.65), arrow = arrow(length = unit(0.3, "inches")), color = "blue")
```

Is the median vaccination rate for NC schools above the CDC recommended target?

```r
nc_vacc_rates |>
    filter(state == 'North Carolina', overall >= 0) |>
    summarize (median_vacc_rate_NC = median(overall))
```
